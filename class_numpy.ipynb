# 1. 넘파일 다차원 배열(ndarray) 객체 생성
- import numpy as np
- 리스트을 이용하여 다차원 배열 생성
- np.array(리스트)
  - 리스트 → 배열로 바꾸고 싶을 때
  - 단순 저장 → 수학 연산이 필요할 때
  - 데이터 분석/머신러닝 모델에 데이터를 입력할 때
- 속성 :
  - ndim : 배열 축 혹은 차원의 개수
  - shape : 배열의 형상을 (n, m) 형식의 튜플형으로 표시
  - size :
    - 배열 원소의 개수
    - (m,n) 형상 배열의 size = m*n
  - dtype : 배열 원소의 자료형을 기술
  - astype : 배열 원소 형 변환

```python
# 1. 리스트를 이용하여 1차원 배열 생성
import numpy as np
data = [10, 20, 30]
print(data)
print("="*100)
arr = np.array(data) # 1차원 배열 반환
print(arr)
print(type(arr))
print(arr.ndim) # 1
print(arr.shape) # (3,)
print(arr.size) # 3
print(arr.dtype) # int64
```

```python
# 1. 중첩리스트를 이용하여 2차원 배열 생성
# 머신러닝에서 행은 샘플의 개수, 열은 각 샘플이 지니고 있는 특징의 개수를 표현
import numpy as np
# 2행 3열 구조로 2명의 학생에 대한 3과목의 점수를 기록-> 넘파이 2차원 배열로 생성
data = [[90, 80, 100], [70, 90, 90]]
print(data)
arr_2d = np.array(data, dtype="float64") # 2d array
print(arr_2d)
print(type(arr_2d))
print(arr_2d.ndim) # 2
print(arr_2d.shape) # (2(학생),3(과목))
print(arr_2d.size) # 2*3 = 6
print(arr_2d.dtype) # float64
```

# 2. 넘파이 다차원 배열 간 사칙 연산

# 6. 다차원 배열의 지정된 축 방향으로 원소들을 뒤집기
  - np.flip(배열객체, axis=)

```python
# (4,5) 구조의 이미지 데이터을 생성 => 상하, 좌우 반전시켜 새로운 이미지 만들기
import numpy as np
# 데이터 생성
img = np.arange(20).reshape((4,-1)) # (4,5)
print(img)

# (1) 좌우 반전
h_flip = np.flip(img, axis = 1)
print(h_flip)

# (2) 상하 반전 (axis = 0)
v_flip = np.flip(img, axis = 0)
print(v_flip)

# (3) 전체 반전 (상하좌우 반전)
all_flip = np.flip(img)
print(all_flip)
```

# 8. 넘파이 다차원 배열의 인덱싱을 이용한 슬라이싱
- 다차원 배열에서 작은 배열을 추출
- 인덱싱을 이용하여 슬라이싱 범위 지정
- 두 축을 이용해서 특정한 행과 열만 별도로 적용하여 추출함
  - axis 0 ,  axis 1
  - arr[] ; axis 0 사용
  - arr[:, :] ;  두 개의 축을 동시에 사용할 때는 쉼표로 구분
  - arr[: :step] ; axis 0을 이용하여 간격을 두고 추출
  - arr[::step, ::step]; 두 개의 축을 이용하여 간격을 두고 추출


```python
# Ex : 3차원 배열 슬라이싱
import numpy as np

# 데이터 생성 : 60개 원소로 구성된 (3,4,5) 형태의 3차원 배열 생성 (3층, 4행, 5열)
tensor = np.arange(60).reshape((3,4,5))
print(tensor)
print("="*100)

# (1) 특정 층 전체 가져오기 (첫번째 층)
print(tensor[0, :, :]) # tensor(0)과 동일 => shape : (4,5), size = 20
print("="*100)

# (2) 모든 층에서 특정 '행'만 가져오기 (모든 층의 첫번째 행)
print(tensor[:, 0, :]) # shape : (3,5), size = 15
print("="*100)

# (3) 모든 층에서 왼쪽 벽면(첫 번째 열)만 수직으로 잘라내기
print(tensor[:, :, 0]) # shape : (3,4), size = 12

# (4) 정가운데 부분만 즉, 큐브형태로 도려내기
print(tensor[1:3, 1:3, 1:4]) # 1-2층의 1~2행에서 1~3열, shape : (2,2,3)












```

# 9.  넘파이 다차원 배열의 슬라이싱을 이용한 논리인덱싱
- 특정한 조건을 만족하는 배열의 원소값만을 추출하여 1차원 배열 객체로 반환
- 브로드캐스팅하여 병렬 연산 처리


1. 주어진 배열을 이용하여 조건식을 만족하는 동일한 형상을 가지는 불리언 배열을 생성

2.   주어진 배열에 불리언 배열을 인덱스로 지정

- cf. np.where(조건식, x, y)
  - 주어진 배열의 구조는 유지하면서 조건식에 맞는 원소를 추출하는 기법
  - if ~ else 동작
  - x는 조건식이 True인 경우 처리
  - y는 조건식이 False인 경우 처리




```python
import numpy as np

# 2차원 배열 데이터 생성
arr = np.arange(100).reshape(5,-1) # (5,20)
print(arr, "\n")

# Ex1 : 주어진 배열에서 짝수들만 추출하여 1차원 배열로 생성
# (1) 불리언 배열
arr_bool = arr % 2 == 0 # (5,20)
print(arr_bool)

# (2) 불리언 언덱싱
arr_even = arr[arr_bool]
print(arr_even) # 1차원 배열

# Ex2 : 주어진 배열에서 20보다 큰 요소들만 추출하여 1차원 배열로 생성
arr_greaterthan_20 = arr[arr > 20]
print(arr_greaterthan_20)

# Ex3 : 주어진 배열에서 20보다 큰 요소들만 추출
arr_fliter = np.where(arr > 20, arr, 0)
print(arr_fliter)
```

# 13.난수을 이용한 다차원 배열 만들기

  - randn(개수) - 표준정규분포의 난수 표본 추출
  - randint(최소값,최대값,개수)- 주어진 범위내의 정수 난수 생성. 최대값은 미포함

  - permutation() - 배열을 임의의 순서로 섞음


```python
# 주가 수익률 그래프 시뮬레이션 (평균 0.001, 표준편차 0.02 인 난수 1000개)
# 의미: 이런 문제의 불확싱성을 난수로 생성하여 향후 1000일 간의 주가를 예측해보기
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import warnings # 경고 메세지 숨기기
import os

# 1. 환경 설정 (경고 차단 및 한글 폰트)
warnings.filterwarnings('ignore')
font_path = '/usr/share/fonts/truetype/nanum/NanumBarunGothic.ttf'
if not os.path.exists(font_path):
    !sudo apt-get install -y fonts-nanum > /dev/null
font_prop = fm.FontProperties(fname=font_path)

# 2. 데이터 준비: 누적 수익률 계산 로직
np.random.seed(42) # 재현성 보장
daily_values = np.random.normal(0.001,0.02,1000)
price = np.cumprod(1 + daily_values) # price.max().price.min() 이용
print(price)

# 2. simuiation : line graph
plt.figure(figsize = (14,5))
plt.plot(np.arange(1000), price, label = "주가 수익률", color = 'r')
plt.title("주가 수익률 시뮬레이션", fontproperties = font_prop, fontsize = 15)
plt.xlabel("일수(days)", fontproperties = font_prop)
plt.ylabel("주가(price)", fontproperties = font_prop)
plt.xticks(np.arange(0,1001,100))
plt.yticks(np.arange(0, 3.7, 0.5))
plt.legend(prop = font_prop)
plt.show()
```

# 14. 다차원 배열 프로그래밍
- 벡터화 연산 (데이터병렬 처리)
  - 배열 간 연산
  - 리덕션
    - np.sum(arr, axis=)
    - np.mean(arr, axis=)
    - arr.sum(axis=0)
    - arr.sum(axis=1)
    - arr.sum(axis=(0,1))
  - 조건 리덕션 원리 (Boolean to Integer)
      - NumPy에서 리덕션이 강력한 이유는 **논리값(True/False)**을 연산하는 순간 **숫자(1/0)**로 자동 변환.

      - np.sum(condition) : 조건이 참(True)인 개수를 구함

      - np.mean(condition) : 조건이 참(True)인 비율을 구함 (예: 0.7이면 70%가 참)

- np.hstack(튜플 또는 리스트)
- np.column_stack(리스트)


```python
#-------------------
# Ex1 1차원 배열간 결합
#-------------------
import numpy as np
# 데이터 생성
a = np.array([1,2,3]) # (3,)
b = np.array([4,5,6]) # (3.)

# 세로 쌓기 (수직방향) -> 행 추가 해서 2차원 배열
res_stack = np.vstack((a,b))
#print(res_stack) # (2,3)

# 가로로 붙이기 (수평방향) -> 열 추가해서 1차원 배열
res_stack = np.hstack((a,b))
#print(res_stack) # (6,)
res_concat = np.concatenate((a,b), axis = 0)
#print(res_concat) # (6,)

# 주의 : column_stack() -> 2차원 배열
res_col = np.column_stack((a,b))
#print(res_col) # (3,2)

#-------------------
# Ex2 형태가 다른 배열간 결합
#-------------------
arr_1 = np.arange(1,4) # (3,)
arr_2 = np.arange(4,10).reshape(2,3) # (2,3)
print(arr_1)
print()
print(arr_2)
print()
# 벙법1
# (1) arr_1 -> (1,3) - 열의 개수를 일치
arr_1_2d = arr_1.reshape(-1,3)
print(arr_1_2d)
print()
combined_2d = np.concatenate((arr_1_2d, arr_2), axis = 0)
print(combined_2d)
print()
# 방법2
v_combined = np.vstack((arr_1, arr_2))
print(v_combined)
```
